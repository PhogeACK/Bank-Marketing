### Feature Engineering

#### Encoding categorial features:

```python
df_encoded = df.copy() #creating a copt of the original df
```

#### One-Hot Encoding for each categorian feature that don't have natural order to it's values:

```python
from sklearn.preprocessing import OneHotEncoder

one_hot_features = ['job', 'marital', 'contact', 'month', 'day_of_week', 'poutcome']

encoder = OneHotEncoder(drop='first', sparse_output=False)
encoded_array = encoder.fit_transform(df_encoded[one_hot_features])

encoded_df = pd.DataFrame(encoded_array, columns=encoder.get_feature_names_out(one_hot_features))

df_encoded = pd.concat([df_encoded.drop(columns=one_hot_features), encoded_df], axis=1)
```

#### Ordianl Encoding for 'Education' feature:

```python
# Ordinal encoding for 'education' feature
education_order = {
    'illiterate': 1,
    'basic.4y': 2,
    'basic.6y': 3,
    'basic.9y': 4,
    'high.school': 5,
    'professional.course': 6,
    'university.degree': 7,
    'unknown': 8
}

# Apply the ordinal encoding to 'education' column
df_encoded['education_encoded'] = df_encoded['education'].map(education_order)
df_encoded = df_encoded.drop(columns=['education'])

df_encoded
```

#### Encoding the boolian faetures:

```python
# Clean the columns by stripping spaces and converting to lowercase (for consistency)
df_encoded['default'] = df_encoded['default'].str.strip().str.lower()
df_encoded['housing'] = df_encoded['housing'].str.strip().str.lower()
df_encoded['loan'] = df_encoded['loan'].str.strip().str.lower()

# Check the unique values in the columns to make sure they match the expected ones
print(df_encoded['default'].unique())
print(df_encoded['housing'].unique())
print(df_encoded['loan'].unique())

# Create dictionaries for encoding the boolean-like features with 'unknown'
boolean_mapping = {
    'yes': 1,
    'no': 0,
    'unknown': -1  
}

# Apply the mapping to 'default', 'housing', and 'loan' features and create new encoded columns
df_encoded['default_encoded'] = df_encoded['default'].map(boolean_mapping)
df_encoded['housing_encoded'] = df_encoded['housing'].map(boolean_mapping)
df_encoded['loan_encoded'] = df_encoded['loan'].map(boolean_mapping)

# Drop the original columns after encoding
df_encoded = df_encoded.drop(columns=['default', 'housing', 'loan'])
```

#### Encoding the Target Variable y:

```python
# Create a dictionary for encoding the target variable 'y'
target_mapping = {
    'yes': 1,
    'no': 0
}

# Apply the mapping to the 'y' column
df_encoded['y'] = df_encoded['y'].map(target_mapping)

df_encoded['y'].value_counts()
```

#### Dropping 'duration' feature:
##### 'duration' contains future information (data leakage) that the model wouldn't have in a real-world scenario.
##### Dropping it ensures that your model is trained with only the features that would be available at the time of prediction, resulting in a more reliable and realistic model.

```python
# Drop the 'duration' feature since it should not be included for realistic predictions
df_encoded = df_encoded.drop(columns=['duration'])
```

#### Treating Pdays:
##### Pdays will be a categorial feature from now on. The value 999 will be changed to -1.

```python
# Correct mapping for 'pdays'
df_encoded['pdays'] = df_encoded['pdays'].replace(999, -1)

# Check if any NaN values exist after the replacement
print(df_encoded['pdays'].isna().sum())  # This should return 0 if no NaNs remain
```
